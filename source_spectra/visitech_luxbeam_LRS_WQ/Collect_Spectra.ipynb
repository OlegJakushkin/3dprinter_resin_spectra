```python
import seabreeze.spectrometers as sb
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
devices = sb.list_devices()
spec = sb.Spectrometer(devices[0])
```

```python
spec.integration_time_micros(8000)
wl = spec.wavelengths()

for i in range(20):
    spectra = spec.intensities()
    result =np.transpose(np.array([wl, spectra]))

    np.savetxt('light_engine_spectra_' + str(i+1) + '.csv', result, delimiter=',', header='wavelength(nm),spectra')

    plt.plot(wl, spectra)
    plt.xlim(350,450)
    plt.show()
```

```python
spec.close()
```
